```python
# Scatterplot comparing budget and gross

plt.scatter(x=df["budget"], y=df["gross"])
plt.title("Budget vs Gross Earnings")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.show()
```

```python
# Plotting budget vs gross using seaborn to find correlation
sns.regplot(x="budget", y="gross", data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Looking at the correlation data
# Correlation methods = pearson, kendall, spearman
df.corr(method="pearson")
```

```python
# High correlation between budget and gross
# Heatmap of default numerical values

correlation_matrix = df.corr(method="pearson")
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# To see correlation between non-numeric features we need to digitize all of the non-numerical columns.
df_digitized = df

for col_name in df_digitized.columns:
    if(df_digitized[col_name].dtype == "object"):
        df_digitized[col_name] = df_digitized[col_name].astype("category")
        df_digitized[col_name] = df_digitized[col_name].cat.codes
df_digitized.head()
```

```python
# Heatmap of all features
correlation_matrix = df_digitized.corr(method="pearson")
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
df_digitized.corr()
```

```python
correlation_matrix = df_digitized.corr()
correlation_pairs = correlation_matrix.unstack()
# pd.set_option('display.max_rows', None)
correlation_pairs
```

```python
sorted_pairs = correlation_pairs.sort_values()
# pd.set_option('display.max_rows', None)
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```
